```python
# Create tensors with gradient tracking enabled
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

# Perform operations
z = x**2 + y**3
print(f"x = {x.item()}")
print(f"y = {y.item()}")
print(f"z = x² + y³ = {z.item()}")

# Compute gradients
z.backward()

# Access gradients
print(f"\ndz/dx = 2x = {x.grad.item()}")
print(f"dz/dy = 3y² = {y.grad.item()}")
```

```python
# Create a simple computational graph
a = torch.tensor([2.0, 3.0], requires_grad=True)
b = torch.tensor([6.0, 4.0], requires_grad=True)

# Operations
c = a + b
d = c * b
e = d.sum()

print("a:", a)
print("b:", b)
print("c = a + b:", c)
print("d = c * b:", d)
print("e = sum(d):", e)

# Compute gradients
e.backward()

print("\nGradients:")
print("de/da:", a.grad)
print("de/db:", b.grad)
```

## 5. Building Neural Networks

PyTorch provides the `nn` module for building neural networks. Let's create a simple feedforward neural network.

```python
# Define a simple neural network class
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        # Define layers
        self.fc1 = nn.Linear(input_size, hidden_size)  # First fully connected layer
        self.relu = nn.ReLU()  # Activation function
        self.fc2 = nn.Linear(hidden_size, output_size)  # Output layer

    def forward(self, x):
        # Define forward pass
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Create an instance of the network
input_size = 10
hidden_size = 20
output_size = 5

model = SimpleNN(input_size, hidden_size, output_size)
print("Neural Network Architecture:")
print(model)

# Test the network with random input
sample_input = torch.randn(1, input_size)
output = model(sample_input)
print(f"\nInput shape: {sample_input.shape}")    #
print(f"Output shape: {output.shape}")
print(f"Output: {output}")
```

```python
# Generate synthetic regression data
# y = 3x + 2 + noise

torch.manual_seed(42)  # For reproducibility
np.random.seed(42)

# Generate 100 samples
X = torch.randn(100, 1) * 10  # Input features
y = 3 * X + 2 + torch.randn(100, 1) * 2  # Target with noise

# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X.numpy(), y.numpy(), alpha=0.5)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Regression Data')
plt.grid(True)
plt.show()

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

```python
# Define a simple linear regression model
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1, 1)  # 1 input, 1 output

    def forward(self, x):
        return self.linear(x)

# Create model instance
model = LinearRegressionModel()
print("Model:")
print(model)

# Print initial parameters
print("\nInitial parameters:")
for name, param in model.named_parameters():
    print(f"{name}: {param.data}")
```

```python
# Define loss function and optimizer
criterion = nn.MSELoss()  # Mean Squared Error Loss
optimizer = optim.SGD(model.parameters(), lr=0.01)  # Stochastic Gradient Descent

# Training loop
num_epochs = 100
losses = []

for epoch in range(num_epochs):
    # Forward pass
    predictions = model(X)
    loss = criterion(predictions, y)

    # Backward pass and optimization
    optimizer.zero_grad()  # Clear previous gradients (prevents gradient accumulation)
    loss.backward()  # Compute gradients
    optimizer.step()  # Update parameters

    # Store loss for plotting
    losses.append(loss.item())

    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Print final parameters
print("\nFinal parameters:")
for name, param in model.named_parameters():
    print(f"{name}: {param.data}")
print("\n(Expected: weight ≈ 3.0, bias ≈ 2.0)")
```

```python
# Plot training loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Time')
plt.grid(True)

# Plot predictions vs actual data
plt.subplot(1, 2, 2)
with torch.no_grad():  # Disable gradient computation for inference
    predicted = model(X).numpy()

plt.scatter(X.numpy(), y.numpy(), alpha=0.5, label='Actual Data')
plt.plot(X.numpy(), predicted, color='red', linewidth=2, label='Predicted Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Model Predictions')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Generate synthetic classification data
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# Generate data
X_np, y_np = make_moons(n_samples=1000, noise=0.2, random_state=42)

# Split into train and test sets
X_train_np, X_test_np, y_train_np, y_test_np = train_test_split(
    X_np, y_np, test_size=0.2, random_state=42
)

# Convert to PyTorch tensors
X_train = torch.FloatTensor(X_train_np)
X_test = torch.FloatTensor(X_test_np)
y_train = torch.LongTensor(y_train_np)
y_test = torch.LongTensor(y_test_np)

# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X_train_np[:, 0], X_train_np[:, 1], c=y_train_np, cmap='viridis', alpha=0.6)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Binary Classification Data')
plt.colorbar(label='Class')
plt.show()

print(f"Training samples: {len(X_train)}")
print(f"Test samples: {len(X_test)}")
```

```python
# Define a neural network for binary classification
class BinaryClassifier(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(BinaryClassifier, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, 2)  # 2 output classes

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        return x

# Create model
classifier = BinaryClassifier(input_size=2, hidden_size=16)
print("Classification Model:")
print(classifier)

# Define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(classifier.parameters(), lr=0.01)

```

```python
# Training loop
num_epochs = 200
train_losses = []
train_accuracies = []

for epoch in range(num_epochs):
    # Forward pass
    outputs = classifier(X_train)
    loss = criterion(outputs, y_train)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Calculate accuracy
    _, predicted = torch.max(outputs.data, 1)
    accuracy = (predicted == y_train).sum().item() / len(y_train)

    train_losses.append(loss.item())
    train_accuracies.append(accuracy)

    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {accuracy:.4f}')
```

```python
# Evaluate on test set
with torch.no_grad():
    test_outputs = classifier(X_test)
    _, test_predicted = torch.max(test_outputs.data, 1)
    test_accuracy = (test_predicted == y_test).sum().item() / len(y_test)

print(f'Test Accuracy: {test_accuracy:.4f}')

# Plot training metrics
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(train_losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(train_accuracies)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')
plt.grid(True)

# Visualize decision boundary
plt.subplot(1, 3, 3)
h = 0.02  # Step size in mesh
x_min, x_max = X_test_np[:, 0].min() - 0.5, X_test_np[:, 0].max() + 0.5
y_min, y_max = X_test_np[:, 1].min() - 0.5, X_test_np[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

with torch.no_grad():
    Z = classifier(torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()]))
    _, Z = torch.max(Z, 1)
    Z = Z.reshape(xx.shape).numpy()

plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X_test_np[:, 0], X_test_np[:, 1], c=y_test_np, cmap='viridis', edgecolors='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary')

plt.tight_layout()
plt.show()
```

```python
# Make predictions on new data
new_data = torch.FloatTensor([[0.5, 0.5], [-0.5, 0.5], [1.0, -1.0]])

# Set model to evaluation mode
classifier.eval()

with torch.no_grad():
    predictions = classifier(new_data)
    probabilities = torch.softmax(predictions, dim=1)
    _, predicted_classes = torch.max(predictions, 1)

print("New Data Points:")
print(new_data)
print("\nPredicted Classes:")
print(predicted_classes)
print("\nClass Probabilities:")
print(probabilities)
```
